# Get Polish given names

```python
import pandas as pd

df = pd.read_html('https://en.wiktionary.org/wiki/Appendix:Polish_given_names')
```

```python
data_masc = df[1]['Masculine names'].drop(['Common diminutives'], axis=1)
data_masc
```

```python
data_fem = df[2]['Feminine names'].drop(['Common diminutives'], axis=1)
data_fem
```

```python
df = pd.DataFrame({'Polish': data_masc['Polish name'].to_list() + data_fem['Polish name'].to_list(),
              'English': data_masc['English equivalent'].to_list() + data_fem['English equivalent'].to_list()})
```

```python
df
```

# Get Polish Toponymns

```python
!pip uninstall tabula
!python -m pip install tabula-py
```

```python
import tabula

dfs = tabula.io.read_pdf('/content/Toponymic_Guidelines_of_Poland_2002.pdf', pages='all', multiple_tables=True)
```

```python
data = dfs[17:24]
```

```python
d1 = pd.DataFrame(data[0])
d2 = pd.DataFrame(data[1])
print(d1.columns)
d2.columns = d1.columns
d1.append(d2)
```

```python
d1 = pd.DataFrame(data[0])
cols = d1.columns

for d in data[1:]:
  d = pd.DataFrame(d)
  d.columns = cols
  d1 = pd.concat([d1, d])

```

```python
d1.reset_index(drop=True, inplace=True)
```

```python
d1
```

```python
df
```

# Combine 2 datasets

```python
df1 = d1[['Polish', 'English']]
```

```python
df_final = pd.concat([df, df1])
```

```python
df_final.reset_index(drop=True, inplace=True)
```
